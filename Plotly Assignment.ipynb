Q3. Using the tips dataset in the Plotly library, Plot a histogram for x= "sex" and y="total_bill" column in
the tips dataset. Also, use the "smoker" column with the pattern_shape parameter and the "day"
column with the color parameter.

```python
fig = go.Figure()
fig.add_trace(
    go.Histogram(
        x=tips['sex'],
        y=tips['total_bill'],
        marker=dict(color='blue'),  # Specify a single color for all bars
        name='sex vs total_bill'
    )
)
shape_mapping = {
    'Yes': 'circle',
    'No': 'square'
}
shapes = [shape_mapping[smoker] for smoker in tips['smoker']]
color_mapping = {
    'Thur': 'red',
    'Fri': 'green',
    'Sat': 'blue',
    'Sun': 'purple'
}
colors = [color_mapping[day] for day in tips['day']]
fig.add_trace(
    go.Scatter(
        x=tips['sex'],
        y=tips['total_bill'],
        mode='markers',
        marker=dict(color=colors, symbol=shapes),
        name='smoker vs total_bill'
    )
)
fig.update_layout(
    title='Histogram and Scatter Plot',
    xaxis=dict(title='Sex'),
    yaxis=dict(title='Total Bill')
)

fig.show()
```

Q5. What is Distplot? Using Plotly express, plot a distplot.

A distplot, short for distribution plot, is a data visualization that displays the distribution of a univariate dataset, typically in the form of a histogram along with a kernel density estimate (KDE) curve. It provides a way to visualize the underlying probability distribution of a continuous variable.

```python
pip install plotly
```

```python
import plotly.express as px

data = [1.2, 1.5, 1.6, 1.8, 2.0, 2.1, 2.2, 2.3, 2.5, 2.5, 2.7, 2.8, 3.0, 3.1, 3.2, 3.5]

fig = px.histogram(data, x=data, nbins=10, histnorm='probability density', marginal='rug', title="Distplot Example")

fig.show()
```
