Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
# Note: Also add title, xlabel, and ylabel to the plot.
```

A scatter plot is a type of data visualization that is used to display the relationship between two continuous variables. In a scatter plot, each data point is represented as a single point on a two-dimensional graph, with one variable plotted on the x-axis and the other variable on the y-axis.

```python
plt.xlabel("X- axis")
plt.ylabel("Y- axis")
plt.title("Scatter Plot")
plt.plot(x,y)
plt.show()
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:

```python
import numpy as np
# For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
# For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
# For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
# For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])
```

The subplot() function in Matplotlib is used to create a grid of subplots within a single figure.The primary reasons for using the subplot() function are as follows:
Multiple Plots in One Figure,Complex Visualizations,Comparison of Data,Custom Layouts

```python
x = np.array([0, 1, 2, 3, 4, 5])
```

```python
y = np.array([0, 100, 200, 300, 400, 500])
plt.subplot(2, 2, 1)
plt.plot(x, y, label='Line 1', c='b')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot 1')

y = np.array([50, 20, 40, 20, 60, 70])
plt.subplot(2, 2, 2)
plt.plot(x, y, label='Line 2', c='y')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot 2')

y = np.array([10, 20, 30, 40, 50, 60])
plt.subplot(2, 2, 3)
plt.plot(x, y, label='Line 3',c='r')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot 3')

y = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(2, 2, 4)
plt.plot(x, y, label='Line 4', c='#A64D13')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot 4')

plt.legend()
plt.tight_layout()
plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

A bar plot (or bar chart) is a common type of data visualization used to represent categorical data in a graphical form.

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.bar(company,profit,color='teal', edgecolor='black')
plt.xlabel("company")
plt.ylabel("profit")
plt.title("bar plot")
plt.show()
```

```python
plt.barh(company,profit,color='red', edgecolor='teal')
plt.ylabel("company")
plt.xlabel("profit")
plt.title("bar plot")
plt.show()
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

A box plot is a graphical representation of the distribution of a dataset.It is particularly useful for visualizing the spread, central tendency, and presence of outliers in a dataset. 

```python
data = [box1, box2]
plt.boxplot(data, labels=['Box 1', 'Box 2'])
plt.xlabel('Box')
plt.ylabel('Values')
plt.title('Box Plot Example')
plt.show()
```
